```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("SUS_vragenlijst.csv")

print(df.head())
```

```python
# ID-kolom hernoemen (handig)
df = df.rename(columns={
    "Wat is je participant ID? ": "id"
})

df["id"] = df["id"].astype(str).str.strip().str.upper()
```

```python
# SUS-vragen in juiste volgorde
sus_questions = [
    "Ik denk dat ik deze website vaak zou willen gebruiken",
    "Ik vond deze website onnodig complex",
    "Ik vond deze website gemakkelijk te gebruiken",
    "Ik denk dat ik ondersteuning nodig zou hebben om deze website te kunnen gebruiken",
    "Ik vond dat de verschillende functies van deze website goed op elkaar waren afgestemd",
    "Ik vond dat er te veel inconsistentie in deze website zat",
    "Ik kan me voorstellen dat de meeste mensen deze website heel snel zouden leren gebruiken",
    "Ik vond deze website erg omslachtig/onhandig in gebruik",
    "Ik voelde me erg zelfverzekerd tijdens het gebruik van deze website",
    "Ik moest veel leren voordat ik met deze website aan de slag kon"
]

# Positieve en negatieve item-indexen (0-based)
positive_idx = [0, 2, 4, 6, 8]
negative_idx = [1, 3, 5, 7, 9]
```

```python
# Scores berekenen
for i, col in enumerate(sus_questions):
    if i in positive_idx:
        df[f"Q{i+1}_score"] = df[col] - 1
    else:
        df[f"Q{i+1}_score"] = 5 - df[col]

# SUS-score
score_cols = [f"Q{i}_score" for i in range(1, 11)]
df["SUS"] = df[score_cols].sum(axis=1) * 2.5
```

```python
# zelf samengestelde groepen
groep_A_ids = ["J1", "J3", "E1", "J4", "E4", "T2", "J6", "E5", "S1", "S3", "S5"]
groep_B_ids = ["J2", "E2", "E3", "J5", "T1", "J7", "S2", "S4", "T3", "T4"]

df["groep"] = "Onbekend"  # default

df.loc[df["id"].isin(groep_A_ids), "groep"] = "Groep A"
df.loc[df["id"].isin(groep_B_ids), "groep"] = "Groep B"
```

```python
df.groupby("groep")["SUS"].median()
```
